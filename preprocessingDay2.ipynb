## **Data cleaning and Encoding**

```python
#Check for missing values and unique entries in categorical columns
missig_val=data.isnull().sum()
uniq_val={col:data[col].unique() for col in ['Department','salary']}
missig_val,uniq_val
```

```python
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder

# One-hot encoding for 'Department'
# The 'sparse' argument has been replaced with 'sparse_output'
department_encoder = OneHotEncoder(sparse_output=False, drop='first')  # Drop first to avoid multicollinearity
department_encoded = department_encoder.fit_transform(data[['Department']]) # Changed hr_data to data
department_encoded_df = pd.DataFrame(department_encoded, columns=department_encoder.get_feature_names_out(['Department']))

# Ordinal encoding for 'salary'
salary_encoder = OrdinalEncoder(categories=[['low', 'medium', 'high']])
data['salary_encoded'] = salary_encoder.fit_transform(data[['salary']]) # Changed hr_data to data

# Combine the encoded 'Department' data back into the dataset and drop original columns
hr_data_encoded = pd.concat([data, department_encoded_df], axis=1).drop(columns=['Department', 'salary']) # Changed hr_data to data

# Display the first few rows of the processed dataset
hr_data_encoded.head()
```
